```python
#Zadanie 3. Napisz klasę Vector2D, która będzie reprezentować wektor dwuwymiarowy. Klasa powinna zawierać metody magiczne
#__add__ oraz __radd__, aby umożliwić dodawanie dwóch wektorów 2D oraz dodawanie krotek (x, y) do wektorów.
class Vector2D:
    def __init__(self, x, y):
        
        # Inicjalizacja obiektu Vector2D z wartościami x i y.
        self.x = x
        self.y = y
    
    def __add__(self, other):
        
        # Przeciążenie operatora dodawania (+) dla obiektów Vector2D
        if isinstance(other, Vector2D):
            
            # Jeśli other jest obiektem Vector2D, dodaj jego wartości x i y do bieżącego obiektu
            return Vector2D(self.x + other.x, self.y + other.y)
        elif isinstance(other, tuple) and len(other) == 2:
            
            # Jeśli other jest krotką (tuple) o długości 2, dodaj jej wartości do x i y bieżącego obiektu.
            return Vector2D(self.x + other[0], self.y + other[1])
        else:
            
            # W przypadku, gdy other nie jest obsługiwany, zgłoś błąd TypeError.
            raise TypeError("Unsupported operand type for +")
    
    def __radd__(self, other):
        
        # Przeciążenie operatora dodawania (+) dla obiektu Vector2D i innych typów (np. tuple)
        if isinstance(other, tuple) and len(other) == 2:
            
            # Jeśli other jest krotką (tuple) o długości 2, dodaj jego wartości do x i y bieżącego obiektu
            return Vector2D(self.x + other[0], self.y + other[1])
        else:
            
            # W przypadku, gdy other nie jest obsługiwany, zgłoś błąd TypeError
            raise TypeError("Unsupported operand type for +")
    
    def __str__(self):
        
        # Przeciążenie operatora str(), aby zwrócić czytelną reprezentację obiektu Vector2D jako string.
        return f"({self.x}, {self.y})"
    
    
# Przykładowe użycie klasy Vector2D:
v1 = Vector2D(0.5, 2)
v2 = Vector2D(3, 4)
v3 = v1 + v2
print(v3)  # Wydrukuj wynik dodawania v1 i v2: (3.5, 6)

v4 = v1 + (5, 6)
print(v4)  # Wydrukuj wynik dodawania v1 i tuple (5.5, 8)

v5 = (7, 8) + v1
print(v5)  # Wydrukuj wynik dodawania tuple (7.5, 10) i v1
```

```python
#Zadanie 4. Napisz klasę Polynomial, która będzie reprezentować wielomian jednej zmiennej. Klasa powinna zawierać 
#metody magiczne __add__, __radd__, __getitem__ oraz __setitem__, aby umożliwić dodawanie dwóch wielomianów, 
#dodawanie liczby rzeczywistej do wielomianu, uzyskiwanie współczynników wielomianu oraz ustawianie wartości współczynników.
class Polynomial:
    def __init__(self, coefficients):
        
        # Konstruktor przyjmuje listę współczynników wielomianu
        self.coefficients = coefficients
    
    def __add__(self, other):
        
        # Metoda umożliwia dodawanie dwóch wielomianów lub dodawanie liczby rzeczywistej do wielomianu
        if isinstance(other, Polynomial):
            
            # Jeśli drugi operand jest wielomianem, dodajemy odpowiednie współczynniki
            max_len = max(len(self.coefficients), len(other.coefficients))
            result_coefficients = [0] * max_len
            for i in range(max_len):
                self_coeff = self.coefficients[i] if i < len(self.coefficients) else 0
                other_coeff = other.coefficients[i] if i < len(other.coefficients) else 0
                result_coefficients[i] = self_coeff + other_coeff
            return Polynomial(result_coefficients)
        elif isinstance(other, (int, float)):
            
            # Jeśli drugi operand jest liczbą rzeczywistą, dodajemy ją do wyrazu wolnego
            result_coefficients = self.coefficients.copy()
            result_coefficients[0] += other
            return Polynomial(result_coefficients)
        else:
            raise TypeError("Unsupported operand type for +")
    
    def __radd__(self, other):
        
        # Metoda umożliwia dodawanie liczby rzeczywistej do wielomianu, gdy liczba jest po lewej stronie operatora +
        if isinstance(other, (int, float)):
            result_coefficients = self.coefficients.copy()
            result_coefficients[0] += other
            return Polynomial(result_coefficients)
        else:
            raise TypeError("Unsupported operand type for +")
    
    def __getitem__(self, index):
        
        # Metoda umożliwia uzyskanie współczynnika wielomianu dla danego indeksu
        if index < len(self.coefficients):
            return self.coefficients[index]
        else:
            return 0
    
    def __setitem__(self, index, value):
        
        # Metoda umożliwia ustawienie wartości współczynnika wielomianu dla danego indeksu
        if index < len(self.coefficients):
            self.coefficients[index] = value
        else:
            self.coefficients.extend([0] * (index - len(self.coefficients) + 1))
            self.coefficients[index] = value
    
    def __str__(self):
        
        # Metoda umożliwia ładne wyświetlenie wielomianu jako string
        terms = []
        for i, coeff in enumerate(self.coefficients):
            if coeff != 0:
                if i == 0:
                    terms.append(repr(coeff))  # Używamy repr() zamiast str()
                elif i == 1:
                    terms.append(f"{coeff}x")
                else:
                    terms.append(f"{coeff}x^{i}")
        return " + ".join(reversed(terms)) if terms else "0"

# Przykłady użycia:

# Tworzenie wielomianów
p1 = Polynomial([2, 1, 3])  # 3x^2 + x + 2
p2 = Polynomial([1, 2, 0, 4])  # 4x^3 + 2x + 1

# Dodawanie wielomianów
p3 = p1 + p2

#Przykład dodawania wielomianów - wydruk
print(p2)  
print(p1) 
print(p3) 
```

```python
#Zadanie 6. Napisz klasę Car, która będzie reprezentować samochód. Klasa powinna zawierać atrybuty instancyjne
#make, model, year, mileage oraz price, inicjalizator, właściwości dla mileage oraz price (umożliwiające odczyt 
#i zapis z walidacją wartości) oraz metody magiczne __str__ i __repr__. Dodatkowo, klasa powinna posiadać metody drive 
#(aktualizującą przebieg samochodu) oraz calculate_depreciation (obliczającą spadek wartości samochodu na podstawie jego 
#przebiegu i wieku).
from datetime import datetime
import datetime
class Car:
        #Inicjalizuje obiekt klasy Car.

        #Parametry:
            #make (str): Marka samochodu
            #model (str): Model samochodu
            #year (int): Rok produkcji samochodu
            #mileage (int lub float): Aktualny przebieg samochodu w kilometrach
            #price (int lub float): Aktualna cena samochodu
        
        #TypeError: Jeśli marka lub model samochodu nie są ciągami znaków
        #ValueError: Jeśli rok produkcji, przebieg lub cena samochodu mają nieprawidłowe wartości
    def __init__(self, make, model, year, mileage, price):
        if not isinstance(make, str) or not isinstance(model, str):
            raise TypeError("Marka i model samochodu muszą być ciągami znaków.")
        if not isinstance(year, int) or year < 1886:
            raise ValueError("Rok produkcji musi być liczbą całkowitą większą lub równą 1886.")
        current_year = datetime.datetime.now().year
        if year > current_year:
            raise ValueError("Rok produkcji nie może być większy od aktualnego roku.")
        if not isinstance(mileage, (int, float)) or mileage < 0:
            raise ValueError("Przebieg musi być nieujemną liczbą.")
        if not isinstance(price, (int, float)) or price < 0:
            raise ValueError("Cena musi być nieujemną liczbą.")
        
        self.make = make
        self.model = model
        self.year = year
        self._mileage = mileage
        self._price = price

    @property
    def mileage(self): #Zwraca aktualny przebieg samochodu.
        return self._mileage

    @mileage.setter
    def mileage(self, value):
        
    #Ustawia nową wartość przebiegu samochodu
    #value (int lub float): Nowa wartość przebiegu samochodu
    #ValueError: Jeśli nowa wartość przebiegu jest nieprawidłowa
        if not isinstance(value, (int, float)) or value < 0:
            raise ValueError("Przebieg musi być nieujemną liczbą.")
        self._mileage = value

    @property
    def price(self): #Zwraca aktualną cenę samochodu
        return self._price

    @price.setter
    def price(self, value):
    #Ustawia nową wartość ceny samochodu
    #value (int lub float): Nowa wartość ceny samochodu
    #ValueError: Jeśli nowa wartość ceny jest nieprawidłowa
        if not isinstance(value, (int, float)) or value < 0:
            raise ValueError("Cena musi być nieujemną liczbą.")
        self._price = value

    def __str__(self): #Zwraca opis samochodu
        return f"{self.make} {self.model} ({self.year}) - [{self.mileage}km, {self.price}zł]"

    def __repr__(self): #Zwraca reprezentację obiektu w postaci stringa.
        return f"Car(make='{self.make}', model='{self.model}', year={self.year}, mileage={self.mileage}, price={self.price})"

    def drive(self, distance):
        if not isinstance(distance, (int, float)) or distance < 0:
            raise ValueError("Dystans musi być nieujemną liczbą.")
        self.mileage += distance

    def calculate_depreciation(self, current_year): #Oblicza aktualną wartość samochodu uwzględniając jego wiek i przebieg
        if not isinstance(current_year, int) or current_year < self.year:
            raise ValueError("Bieżący rok musi być liczbą całkowitą większą lub równą rokowi produkcji samochodu.")
        
        age = current_year - self.year
        depreciation = min(0.05 * age + 0.001 * (self.mileage // 1000), 1.0) 
        #Założenia: Każdy rok kosztuje 5% wartosci, za każdy 1 tys. km samochód traci 0.1% wartości, wartość nie może być ujemna
        depreciated_price = max(self.price * (1 - depreciation), 0)
        return depreciated_price

# Przykłady użycia
try:
    car1 = Car("Toyota", "Camry", 2018, 50000, 25000) #Definiowanie przykładowego samochodu
    print(f"Wydruk 1: Przykład zdefiniowanego obiektu klasy Car: \n {car1}")
    print(f"Wydruk 2: Reprezentacja obiektu Car w postaci łańcucha znaków:\n {repr(car1)}")

    car1.drive(1000) #Dodawanie 1000km do przebiegu
    print(f"Wydruk 3: Przebieg samochodu po dodaniu 1000km:\n {car1.mileage}")

    selling_year = 2025 #Rok w którym planuje sprzedaż
    depreciated_price = car1.calculate_depreciation(selling_year) #Jaka będzie wartość samochodu w roku 2025?
    print(f"Wydruk 4: Zdeprecjonowana wartość samochodu w roku {selling_year} i przejechaniu dodatkowego 1 tys. km wynosi {depreciated_price} zł")
    
    # Próba utworzenia samochodu z nieprawidłowymi danymi
    print("Wydruk 5,6,7,8: Przykłady obsługi błędów:")
    car2 = Car("Ford", "Mustang", 1800, -1000, 20000)
    print(car3)
except (TypeError, ValueError) as e:
    print(f"Błąd: {str(e)}")
    
try:
    car3 = Car(123, "Mustang", 1895, 1000, 20000)
except (TypeError, ValueError) as e:
    print(f"Błąd: {str(e)}")
try:
    car4 = Car("Ford", "Mustang", 1895, -1000, 20000)
except (TypeError, ValueError) as e:
    print(f"Błąd: {str(e)}")
try:
    car5 = Car("Ford", "Mustang", 1895, 1000, "Toyota")
except (TypeError, ValueError) as e:
    print(f"Błąd: {str(e)}")
```

```python
#Zadanie 7. Napisz klasę Samochod, która będzie mieć następujące atrybuty instancyjne:

#marka - marka samochodu
#model - model samochodu
#rok_produkcji - rok produkcji samochodu
#przebieg - przebieg samochodu
#Klasa powinna mieć następujące metody:

#__init__(self, marka, model, rok_produkcji, przebieg) - konstruktor, który będzie inicjalizował atrybuty marka,
#model, rok_produkcji i przebieg
#__str__(self) - metoda magiczna, która będzie zwracać reprezentację napisową obiektu klasy Samochod
#__lt__(self, other) - metoda magiczna, która będzie porównywać dwa samochody po ich przebiegu. 
#Metoda ma zwracać True, jeśli przebieg samochodu self jest mniejszy niż przebieg samochodu other, a w przeciwnym wypadku False.
from datetime import datetime
import datetime

class Samochod:
    def __init__(self, marka, model, rok_produkcji, przebieg):
                #Konstruktor klasy Samochod. Inicjalizuje atrybuty instancyjne
        if not isinstance(marka, str) or not isinstance(model, str):
            raise TypeError("Marka i model samochodu muszą być ciągami znaków.")
        if not isinstance(rok_produkcji, int) or not isinstance(przebieg, int):
            raise TypeError("Rok produkcji i przebieg muszą być liczbami całkowitymi (int).")
        if przebieg < 0:
            raise ValueError("Przebieg nie może być ujemny.")
        if rok_produkcji < 1886:  # pierwszy samochód powstał w 1886 roku
            raise ValueError("Rok produkcji musi być liczbą całkowitą większą lub równą 1886.")
        current_year = datetime.datetime.now().year
        if rok_produkcji > current_year:
            raise ValueError("Rok produkcji nie może być większy od aktualnego roku.")
        # Inicjalizacja atrybutów instancyjnych
        #Parametry:
            #marka (str): Marka samochodu.
            #model (str): Model samochodu.
            #rok_produkcji (int): Rok produkcji samochodu.
            #przebieg (int lub float): Aktualny przebieg samochodu w kilometrach.
        self.marka = marka
        self.model = model
        self.rok_produkcji = rok_produkcji
        self.przebieg = przebieg
    
    def __str__(self):
        #Zwraca reprezentację obiektu w postaci stringa.
        return f"{self.marka} {self.model} ({self.rok_produkcji}), przebieg: {self.przebieg} km"
    
    def __lt__(self, other):
        # Porównuje dwa samochody po ich przebiegu
        return self.przebieg < other.przebieg


# Przykład użycia:
samochod1 = Samochod("Toyota", "Camry", 2020, 50000)
samochod2 = Samochod("Honda", "Civic", 2019, 60000)

print(f"Wydruk 1: Samochód1:\n {samochod1}")  # Toyota Camry (2020), przebieg: 50000 km
print(f"Wydruk 2: Samochód2:\n {samochod2}")  # Honda Civic (2019), przebieg: 60000 km

print(f"Wydruk 3: Czy przebieg Toyoty jest mniejszy niż Hondy?\n Wynik: {samochod1 < samochod2}")  # True
print(f"Wydruk 4: Czy przebieg Hondy jest mniejszy niż Toyoty?\n Wynik: {samochod1 > samochod2}")  # False
```
